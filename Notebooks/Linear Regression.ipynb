# Linear regression with one variable
The dataset used can be found on [Kaggle](https://www.kaggle.com/saxenapriyansh/coursera-machine-learning-andrew-ng#week_1-ex_1.txt) and falls under GPL.

As CEO of a restaurant franchise you are considering different cities for opening a new outlet. Our company already has trucks in various cities and we have data about profits and populations from the cities. A negative value for profit indicates a loss.

Our dataset contains following columns in order:
- populations from the cities divided by 1000
- profits of a food truck in that city divided by 1000

## Loading your dataset

```python
#Imports
import numpy as np;
import matplotlib.pyplot as plt;
import pandas as pd;

#Used for showing inline graphs instead of new window
%matplotlib inline

#Load data
dataset = pd.read_csv('food_trucks.csv');
```

## Data exploration

```python
#Print information about dataset
print(dataset.dtypes);
print("\nSize of dataset: ");
print(dataset.shape);

#Print dataset first 5 rows
print('\n');
print(dataset.head());

#Describe value columns
print('\n');
print(dataset.describe());
```

## Data visualisation

```python
#Split the columns
X = dataset.iloc[:, :-1].values;
y = dataset.iloc[:, 1].values;

#Plot the data
plt.scatter(X,y);
plt.xlabel('Population (x10,000)');
plt.ylabel('Profit (x10,000)');
plt.show();
```

```python
#Split the datset into training and test sets

#Import helper
from sklearn.model_selection import train_test_split;

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0);
```

```python
#Select ML algorithm and train the model

#Import algorithm
from sklearn.linear_model import LinearRegression;

regressor = LinearRegression();
regressor.fit(X_train, y_train);
```

```python
#Make and visualise your prediction
y_pred = regressor.predict(X_test);

# Visualizing the Training set results
plt.scatter(X_train, y_train);
plt.plot(X_train, regressor.predict(X_train), color = 'orange');
plt.title("Training set")
plt.xlabel('Population (x10,000)');
plt.ylabel('Profit (x10,000)');
plt.show();

# Visualizing the Test set results
plt.scatter(X_test, y_test);
plt.plot(X_train, regressor.predict(X_train), color = 'orange');
plt.title("Test set")
plt.xlabel('Population (x10,000)');
plt.ylabel('Profit (x10,000)');
plt.show();
```

```python
#Print the R-squared value
rsquared = regressor.score(X_test,y_test)
print(rsquared * 100,'%')
```
